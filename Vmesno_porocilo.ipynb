**Opis problema:**

Za izdelavo seminarske naloge smo si za problem izbrali prometne nesreče na slovenskih cestah do leta 2014.

Podatke smo razvrstili po naslednih dejavnikih:
- mesecih v letu
- urah v dnevu
- vrsta ceste
- vinjeni udeleženci v prometu

Podatke smo našli na spletni stani https://podatki.gov.si.
Vse analize in ugotovitve bomo izdelali s pomočjo štirih tabel.





```python
import csv
st_nesrec_mesec = []
firstline = True
secondline = True
meseci = {"januar":0, "februar":0, "marec":0, "april":0, "maj":0, "junij":0, "julij":0, "avgust":0, "september":0, "oktober":0, "november":0, "december":0}
with open('nesrece_po_mesecih.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';', quotechar='"')
    for row in spamreader:
        if firstline:
            firstline = False
            continue
        if secondline:
            secondline = False
            continue
        st_nesrec_mesec.append([row[1], row[10], row[19], row[28], row[37], row[46], row[55], row[64], row[73], row[82], row[91], row[100]])
    for row in st_nesrec_mesec:
        meseci["januar"]+=int(row[0])
        meseci["februar"]+=int(row[1])
        meseci["marec"]+=int(row[2])
        meseci["april"]+=int(row[3])
        meseci["maj"]+=int(row[4])
        meseci["junij"]+=int(row[5])
        meseci["julij"]+=int(row[6])
        meseci["avgust"]+=int(row[7])
        meseci["september"]+=int(row[8])
        meseci["oktober"]+=int(row[9])
        meseci["november"]+=int(row[10])
        meseci["december"]+=int(row[11])
#meseci
```

```python
import matplotlib.pyplot as plt
import operator
import numpy as np
```

```python
koncni = meseci.items()
koncni_nesrece = []
for t in koncni:
    koncni_nesrece.append((t[0],t[1]))
    
koncni_poskodovani = meseci_poskodbe.items()
koncni_poskodbe = []
for t in koncni_poskodovani:
    koncni_poskodbe.append((t[0],t[1]))
x1,y1 = zip(*koncni_poskodbe)

x, y = zip(*koncni_nesrece)
plt.figure(figsize=(15, 5))
plt.plot(x,y,label='Število nesreč')
plt.plot(x1,y1,label='Število udeležencev z lahkimi pokšodbami')
plt.legend(loc=1)
#plt.ylabel('št. nesreč')
plt.xlabel('meseci')
plt.title('Podatki od leta 2003 do leta 2014 po mesecih')
plt.show()
```
