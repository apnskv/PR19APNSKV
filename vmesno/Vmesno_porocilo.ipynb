Vse analize in ugotovitve bomo izdelali s pomočjo treh tabel.





```python
import csv
st_nesrec_leto = []
firstline = True
secondline = True
leta = {"2003":0, "2004":0, "2005":0, "2006":0, "2007":0, "2008":0, "2009":0, "2010":0, "2011":0, "2012":0, "2013":0, "2014":0}
with open('nesrece_po_mesecih.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in spamreader:
        if firstline:
            firstline = False
            continue
        if secondline:
            secondline = False
            continue
        st_nesrec_leto.append([row[1], row[10], row[19], row[28], row[37], row[46], row[55], row[64], row[73], row[82], row[91], row[100]])
    for row in st_nesrec_leto:
        leta["2003"]+=int(row[0])
        leta["2004"]+=int(row[1])
        leta["2005"]+=int(row[2])
        leta["2006"]+=int(row[3])
        leta["2007"]+=int(row[4])
        leta["2008"]+=int(row[5])
        leta["2009"]+=int(row[6])
        leta["2010"]+=int(row[7])
        leta["2011"]+=int(row[8])
        leta["2012"]+=int(row[9])
        leta["2013"]+=int(row[10])
        leta["2014"]+=int(row[11])
#meseci
```
